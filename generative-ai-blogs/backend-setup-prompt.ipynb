````python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai.api_key = os.getenv('OPENAI_API_KEY')

def get_completion(messages, model="gpt-3.5-turbo", temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]

system_role= {"role": "system", "content": """
    Act as a Software Developer who code using React and Node.js.
"""}

assistant_role= {"role": "assistant", "content": """
    Act as an assistant to a Software Developer who code using React and Node.js.
"""}
````

````python
repo_init_prompt = {"role": "user", "content": """
    Please help with the below tasks:
    - Node version to use and why?
    - Use app name: "timesheet-service"
    - Project setup
        - How to install node using nvm in macos
        - How to init the node project with typescript as the language.
        - How to setup the Amazon RDS database
        - How to use AWS Secrets Manager to store the secrets
        - How to setup SonarQube for code quality
    - Provide the the package.json code for the below:
        - Add dependencies with the stable versions
            -Use npm as the package manager.
            - Use express.js as the back end web application framework 
            - Use express-session for session-based authentication.
            - Use Istanbul for code coverage
                - Build failure on Code coverage threshold below 95%
                - Generate html report
            - Use Snyk for vulnerability scanning                
                - Set fail on high severity
            - ESLint for static code analysis
                - Build failure on error
                - Generate html report
            - Jest for unit testing            
            - Supertest for Api testing
            - Use swagger for Api documentation             
        - Scripts for the below:
            - build: to build the project
            - start: to start the project
            - dev: to start the project in development mode   
            - test: to run the unit tests
            - coverage: to run the code coverage
            - lint: to run the static code analysis
            - snyk: to run the vulnerability scanning                                            
"""}

messages = [system_role, repo_init_prompt]
response = get_completion(messages)
print(response)

repo_response = {"role": "system", "content": response} 
````
