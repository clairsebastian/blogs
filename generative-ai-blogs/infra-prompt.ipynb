```python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai.api_key = os.getenv('OPENAI_API_KEY')

def get_completion(messages, model="gpt-3.5-turbo", temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]

system_role= {"role": "system", "content": """
    Act as a Software Developer assistant who code using React.js and deploy the application to AWS use AWS Cloudformation and CodePipeline.    

    Key values to use:
    - Application name: timesheet-app

    Output must always be in markdown syntax with proper new lines and indentations.
"""}
```

```python
messages = [system_role, 
{"role": "user", "content": """

Requirements to meet:
- Tag all resources with the tag "Project": "aus-generative-ai-experiment"
- Suffix the aws resources with the resource type. For example, S3 bucket name should be timesheet-app-s3-bucket
- No need to generate cli commands, only cloudformation template is required.
- Do not use example domain, keep the default domain provided by AWS.

Please provide the below
- Explain the steps in detail to deploy a react js application S3 and CloudFront.
- Provide the AWS cloudformation template for the same.
    
"""}]

response = get_completion(messages)
print(response)

cloudformation_response= {"role": "assistant", "content": response}
```

```python
messages = [system_role, cloudformation_response, 
{"role": "user", "content": """
    Task is to create a AWS CodePipeline to build and deploy the react js application to the above S3 created.

    Please provide the below:
    - AWS CloudFormation template to create a AWS CodePipeline that does the below
        - Build the react js application at github.com/aus-generative-ai-experiment/timesheet-app
        - Deploy the react js application to the S3 bucket created above        
    - Also include the AWS CloudFormation code to create the service role required by the pipeline neceassary
"""}]

response = get_completion(messages)
print(response)
```

```python
messages = [system_role, {"role": "user", "content": """

Requirements to meet:
- Tag all resources with the tag "Project": "aus-generative-ai-experiment"
- Suffix the aws resources with the resource type. For example, S3 bucket name should be 'timesheet-service-s3-bucket'
- No need to generate cli commands, only cloudformation template is required.
- Do not use example domain, keep the default domain provided by AWS.

Provide the AWS cloudformation template for the below:
- To Deploy a Node.js application using Elastic Beanstalk.
- The application uses Amazon RDS for the database.
- The application should be deployed to an EC2 instance.
- Install Nodejs and NPM on the launch of EC2 instance.

"""}]

response = get_completion(messages)
print(response)

cloudformation_response= {"role": "system", "content": response}
```
