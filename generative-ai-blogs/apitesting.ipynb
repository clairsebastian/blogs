````python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai.api_key = os.getenv('OPENAI_API_KEY')

def get_completion(messages, model="gpt-3.5-turbo", temperature=0):
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature, 
    )
    return response.choices[0].message["content"]

%store -r product_name
%store -r product_high_level_requirement
%store -r story_with_ui_mockups
%store -r architecture

# system_role= {"role": "system", "content": """
#     Act as a Software Developer who is working on the {product_name} application.

#     Key values to use:
#     - Application name: {product_name}

#     Output must always be in markdown syntax with proper new lines and indentations.
# """}
````
