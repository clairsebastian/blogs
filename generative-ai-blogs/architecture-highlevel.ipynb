````python
import openai
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
openai.api_key = os.getenv('OPENAI_API_KEY')
# print(openai.api_key)
````

````python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
````

````python
%store -r product_name
%store -r product_high_level_requirement
%store -r story_with_ui_mockups

print(product_high_level_requirement)
````

````python
architecture = {
    "software-architecture-pattern":"",
    "system-architecture" :"",    
    "architectural-diagram" :"",
    "test-plan": "",
    "cfr-specification": "",
    "domain-model": "",
    "service-apis": ""
}
````

````python
architecture_pattern_prompt = f""" 
High level Product Features = ```{product_high_level_requirement}```
Epics and User Story list for the application named {product_name}: ```{story_with_ui_mockups}```   
Users who would use the {product_name} application are: ```Employee, Manager, Accountant```

Your task is to perform the following actions: 
1. Recommend a software architecture pattern that defines the detailed structure of the system, including the major \
    components, their interactions, data flow, and the system's overall architecture    
"""
arch_pattern_response = get_completion(architecture_pattern_prompt)
print("\n Solution Architecture Document : \n")
print(arch_pattern_response)

architecture["software-architecture-pattern"] = arch_pattern_response
%store architecture
````

````python
detailed_architecture_prompt = f""" 
High level Product Features = ```{product_high_level_requirement}```
Epics and User Story list for the application named {product_name}: ```{story_with_ui_mockups}```
Software Architecture Pattern = ```{architecture["software-architecture-pattern"]}```
Application users of the {product_name} are: ```Employee, Manager, Accountant```

Your task is to perform the following actions acting as a Software Architect: 
1. Provide a Detailed Design Document about the components of the system and their workflows considering the different users of the application.
2. Provide a Technical Specification Document that describes the technical aspects of the system, including the \
        hardware and software requirements, technology stack, database and development and deployment tools, production environment, security measures to be used etc.
"""

detailed_arch_response = get_completion(detailed_architecture_prompt)
print("\n System Architecture Document : \n")
print(detailed_arch_response)

architecture["system-architecture"] = detailed_arch_response
%store architecture
````

````python
arch_diag_prompt = f""" 
High level Product Features = ```{product_high_level_requirement} ```
Epics and User Story list for the application named {product_name}: ```{story_with_ui_mockups}```
Software Architecture  = ```{architecture["software-architecture-pattern"]} {architecture["system-architecture"]}```
Application Users of the {product_name}  are: ```Employee, Manager, Accountant```

Act as a Software Architect and  perform the following actions: 
1. Analyze the Product Features, Epics and Story List and Software Architecture in detail.
2. Provides a C4 model architectural diagram about the different components of the system.
    
"""
arch_diag_response = get_completion(arch_diag_prompt)
print("\n Solution Architecture Diagram : \n")
print(arch_diag_response)
architecture["architectural-diagram"] = arch_diag_response
%store architecture
````

````python
system_test_prompt = f""" 
High level Product Features = ```{product_high_level_requirement}```
Epics and User Story list for the application named {product_name}: ```{story_with_ui_mockups}```

Your task is to perform the following actions: 
1. Analyze the Product Features, Epics and Story List  in detail and Generate a Detailed Test Plan that recommends :
        a. A Testing strategy for the system features, 
        b. Test Objectives
        c. Types of tests to be performed
        d. Test environment
        e. Test schedule and resources
        f. Test deliverables.
    
"""
sys_test_response = get_completion(system_test_prompt)
print("\n Test Plan : \n")
print(sys_test_response)

architecture["test-plan"] = sys_test_response
%store architecture
````

````python
cross_functional_prompt = f"""

Cross Functional Requirements (CFR) = ```The Cross-functional requirements(CFRs) of a product define the systems operational \
    attributes like performance, availability, reliability, usability, accessibility, audit, logging, recoverability, \
    Capacity, security, localisation, etc. These requirements define how the system should perform.```

High level Product Features = ```{product_high_level_requirement}```
Epics and User Story list for the application named {product_name}: ```{story_with_ui_mockups}```
Software Architecture  = ```{architecture["software-architecture-pattern"]} {architecture["system-architecture"]}```
Application Users of the {product_name}  are: ```Employee, Manager, Accountant```

Below are few examples of CFR:
1. The website pages should load in 3 seconds with the total number of simultaneous users < 5 thousand.
2. The system should be able to handle 20 million users without performance deterioration.
3. The system must perform without failure in 95 percent of use cases during a month.

Your task is to Analyze the Product Features, Epics and Story List and Software Architecture Pattern in detail and \
    generate a Detailed Cross Functional Requirement Documentation for the application.
    
"""
cfr_response = get_completion(cross_functional_prompt)
print("\n Cross functional Requirements : \n")
print(cfr_response)

architecture["cfr-specification"] = cfr_response
%store architecture
````
